```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### One hot representation
# Read all words from sentences -> pass to one_hot with vector size -> covert to list
one_hot_repr=[one_hot(words, voc_size) for words in sentences] 
one_hot_repr

##  For each word its given the index 
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding 
#from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# All sentences has different length. Hence all should have same size. so using pad_sequences to fill space.
sent_length = 8
# padding='pre' adds the 0 at beginning whereas `post` adds into the last.
embedded_docs=pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
## Feature representation
dim=10
```

```python
# Create the model
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
# Train the model
model.compile('adam', 'mse')
```
